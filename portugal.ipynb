```python
import matplotlib.pyplot as plt
import geopandas as gpd
from geopandas import GeoDataFrame
import os
import sys
from pathlib import Path

new_dir =  "C:/Users/Vicky/Desktop/Bewerbung/"
os.chdir(new_dir)  

path_limits = 'Project_Lisboa\Limites_GestaoUrbana.geojson'
limits_urban = gpd.read_file(path_limits)

path_jardim = 'Project_Lisboa\POIArLivre.geojson'
jardim = gpd.read_file(path_jardim) #INF_DESCRICAO

path_artes = 'Project_Lisboa\Artes.geojson'
artes = gpd.read_file(path_artes)  #NOME

path_cinema = 'Project_Lisboa\Cinema.geojson'
cinema = gpd.read_file(path_cinema) #NOME

path_fitness = 'Project_Lisboa\Fitness.geojson'
fitness = gpd.read_file(path_fitness) #NOME

path_museu = 'Project_Lisboa\Museu_Bib.geojson'
museu_bib = gpd.read_file(path_museu) #NOME
```

```python
import folium
import pandas as pd
import geopandas as gpd
import matplotlib.cm as cm
from matplotlib.colors import rgb2hex
import ipywidgets as widgets

# Create a dropdown menu to select the information to display
dropdown_options = ['Jardims', 'Artes', 'Cinema', 'Fitness', 'Museu & Biblioteca']
dropdown_widget = widgets.Dropdown(options=dropdown_options, value=dropdown_options[0], description='Select:')

# Create an initial map centered on Lisbon
map_lisbon = folium.Map(location=[38.7223, -9.1393], zoom_start=12)
display(map_lisbon)  # Display the initial map

def update_map(option): #,map_lisbon
    
    global map_lisbon  # Declare map_lisbon as a global variable
    # Clear previous markers
    map_lisbon.save('map.html')

    # Create a new map instance
    map_lisbon = folium.Map(location=[38.7223, -9.1393], zoom_start=12)

    # Add district boundaries to the map with the customized style
    # Get unique categories in the 'UIT' column
   
    
    # Add district boundaries to the map with the customized style
    # Get unique categories in the 'UIT' column
    categories = limits_urban['UIT'].unique()

    # Get the number of unique categories
    num_categories = len(categories)

    # Define a colormap based on the number of categories
    cmap = cm.get_cmap('Set1', num_categories)

    # Assign colors to each category
    color_dict = {category: rgb2hex(cmap(i)) for i, category in enumerate(categories)}

    # Assign colors to each geometry based on the category
    def style_function(feature):
        category = feature['properties']['UIT']
        color = color_dict[category]
        return {
            'fillColor': color,
            'color': 'black',
            'weight': 1,
            'fillOpacity': 0.6
        }
    folium.GeoJson(limits_urban, name='UIT', style_function=style_function).add_to(map_lisbon)

    if option == 'Jardims':
        data, column_name = jardim, 'INF_DESCRICAO'
        if 'geometry' in data.columns and not data.empty:
            for idx, row in data.iterrows():
                if row.geometry is not None:
                    iframe = folium.IFrame(str(row[column_name]))
                    popup_text = folium.Popup(iframe, min_width=300, max_width=300)
                    folium.Marker(
                        location=[row.geometry.centroid.y, row.geometry.centroid.x],
                        popup=popup_text,
                        icon=folium.Icon(color='green', icon='leaf')
                    ).add_to(map_lisbon)

    elif option == 'Artes':
        data, column_name = artes, 'NOME'
        if 'geometry' in data.columns and not data.empty:
            for idx, row in data.iterrows():
                if row.geometry is not None:
                    iframe = folium.IFrame(str(row[column_name]))
                    popup_text = folium.Popup(iframe, min_width=300, max_width=300)
                    folium.Marker(
                        location=[row.geometry.centroid.y, row.geometry.centroid.x],
                        popup=popup_text,
                        icon=folium.Icon(color='yellow', icon='star')
                    ).add_to(map_lisbon)

    elif option == 'Cinema':
        data, column_name = cinema, 'NOME'
        if 'geometry' in data.columns and not data.empty:
            for idx, row in data.iterrows():
                if row.geometry is not None:
                    iframe = folium.IFrame(str(row[column_name]))
                    popup_text = folium.Popup(iframe, min_width=300, max_width=300)
                    folium.Marker(
                        location=[row.geometry.centroid.y, row.geometry.centroid.x],
                        popup=popup_text,
                        icon=folium.Icon(color='black', icon='play-circle')
                    ).add_to(map_lisbon)

    elif option == 'Fitness':
        data, column_name = fitness, 'NOME'
        if 'geometry' in data.columns and not data.empty:
            for idx, row in data.iterrows():
                if row.geometry is not None:
                    iframe = folium.IFrame(str(row[column_name]))
                    popup_text = folium.Popup(iframe, min_width=300, max_width=300)
                    folium.Marker(
                        location=[row.geometry.centroid.y, row.geometry.centroid.x],
                        popup=popup_text,
                        icon=folium.Icon(color='red', icon='heartbeat')
                    ).add_to(map_lisbon)

    elif option == 'Museu & Biblioteca':
        data, column_name = museu_bib, 'NOME'
        if 'geometry' in data.columns and not data.empty:
            for idx, row in data.iterrows():
                if row.geometry is not None:
                    iframe = folium.IFrame(str(row[column_name]))
                    popup_text = folium.Popup(iframe, min_width=300, max_width=300)
                    folium.Marker(
                        location=[row.geometry.centroid.y, row.geometry.centroid.x],
                        popup=popup_text,
                        icon=folium.Icon(color='purple', icon='book')
                    ).add_to(map_lisbon)

    display(map_lisbon)

# Initialize the map with the 'Jardims' option
update_map('Jardims')

# Add a callback function to update the map when the dropdown value changes
dropdown_widget.observe(lambda change: update_map(change.new), names='value')


# Display the dropdown widget
display(dropdown_widget)
```
